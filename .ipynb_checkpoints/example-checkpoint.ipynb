```python
def stability_cmap(LE):
    vmin = np.min(LE[:,:])
    vmax = np.max(LE[:,:])
    colors1 = plt.cm.PuRd(np.linspace(.25, 1., 1024))[::-1]        
    colors2 = plt.cm.Blues(np.linspace(.25, 1., int(1024*vmax/abs(vmin))))
    colors = np.vstack((colors1, colors2))
    mymap = mcolors.LinearSegmentedColormap.from_list('my_colormap', colors)
    mymap.set_under('white', 1.0)
    return mymap 
```

This example corresponds to the stability calculations in the paper (LINK). The code is fairly general and can be used to analyze any system of coupled 1D oscillators on an arbitrary undirected hypergraph topology.

# Stability calculations

### Defining the node and coupling dynamics:

```python
def self_term(x):
    return (np.sin(x + np.pi/4))**2
    
def dyadic_term_attractive(x):
    return (np.sin(x[1] + np.pi/4))**2 - (np.sin(x[0] + np.pi/4))**2

def dyadic_term_repulsive(x):
    return - (np.sin(x[1] + np.pi/4))**2 + (np.sin(x[0] + np.pi/4))**2
    
def triadic_term_attractive(x):
    return ( np.sin(x[1] + x[2] - 2*x[0]))

def triadic_term_repulsive(x):
    return (- np.sin(x[1] + x[2] - 2*x[0]))

dynamics_terms = {}
dynamics_terms['2'] = {}
dynamics_terms['3'] = {}
dynamics_terms['1'] = self_term
dynamics_terms['2']['attractive'] = dyadic_term_attractive
dynamics_terms['2']['repulsive'] = dyadic_term_repulsive
dynamics_terms['3']['attractive'] = triadic_term_attractive
dynamics_terms['3']['repulsive'] = triadic_term_repulsive
```

### Defining the Jacobian terms.

Specifically, derivatives of the form $\dfrac{\partial F(x_i)}{\partial x_i}$, $\dfrac{\partial G(x_i,x_j,x_k,...)}{\partial x_j}$.

```python
def self_term_J(x):
    return np.sin(2*x + np.pi/2)
        
def dyadic_term_J_attractive(x):
    return np.sin(2*x[1] + np.pi/2)

def dyadic_term_J_replusive(x):
    return - np.sin(2*x[1] + np.pi/2)
        
def triadic_term_J_attractive(x):
    return np.cos(x[1] + x[2] - 2*x[0])

def triadic_term_J_repulsive(x):
    return - np.cos(x[1] + x[2] - 2*x[0])
 
jacobian_terms = {}
jacobian_terms['2'] = {}
jacobian_terms['3'] = {}
jacobian_terms['1'] = self_term_J
jacobian_terms['2']['attractive'] = dyadic_term_J_attractive
jacobian_terms['2']['repulsive'] = dyadic_term_J_replusive
jacobian_terms['3']['attractive'] = triadic_term_J_attractive
jacobian_terms['3']['repulsive'] = triadic_term_J_repulsive
```

### Running the Maximum Lyapunov Exponent calculation. 

An example for specific parameter values $\beta$, $\sigma^{(2)}$, and $\sigma^{(3)}$.

```python
T = 10000 # total number of time steps
T_mle = 5000 # steps used for MLE estimation

beta = 1.8

sigma = {}
sigma['2'] = {}
sigma['3'] = {}
s2 = .1
sigma['2']['attractive'] = s2
sigma['2']['repulsive'] =  s2
s3 = .1
sigma['3']['attractive'] = s3
sigma['3']['repulsive'] = s3

dyn = dynamics.Dynamics(cs, dynamics_terms, jacobian_terms, beta, sigma)
dyn_quotient, dyn_perturbations, norm = dyn.extended_dynamics_evolution(T)

mle = np.sum(norm[-T_mle:])/T_mle
```

The code can be easily run in parallel to analyze the linear stability for a range of parameters. Using the result from running code on a cluster, we plot the stability results:

```python
MLE = pickle.load(open("mle.p", "rb")) 
```

```python
extent=(0, .4, 0, .2)
mymap = stability_cmap(MLE)
plt.imshow(MLE.T, origin='lower', cmap=mymap, extent=extent, aspect=2)

plt.xlabel(r"$\sigma^{(2)}$", fontsize=18)
plt.ylabel(r"$\sigma^{(3)}$", fontsize=18)
clb = plt.colorbar(fraction=0.045, pad=0.04)
clb.ax.set_title('Re(MLE)')
```
